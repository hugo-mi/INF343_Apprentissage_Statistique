```python
from sklearn.base import BaseEstimator, ClassifierMixin
from collections import Counter

class KNNClassifier(BaseEstimator, ClassifierMixin):
    
    """ Homemade kNN classifier class """

    def __init__(self, n_neighbors=3):
        # Complete the mehtod
        self.n_neighbors = n_neighbors

    def fit(self, X, y):
        # Complete the mehtod
        self.X = X
        self.y = y
        return self.X,self.y

    def predict(self, X):
    # Complete the mehtod

        list_dist = list()
        index_list = list()
        class_list = list()

        # 1ère étape : calculer les distances euclidiennes entre le point X et les points du dataset
        for x_value in self.X:
            dist_euc = np.sqrt(np.sum(np.square(x_value-X)))
            list_dist.append(dist_euc)

        # 2ème étape : On ordonne la liste par ordre croissant (plus petite distance à la plus grande distance)
        list_dist_sorted = sorted(list_dist)

        # 3ème étape : On garde les K plus petites distances
        K_min_dist = list_dist_sorted[0:self.n_neighbors]

        # 4 ème étape : On regarde la classe des K points dont la distance à été calculée)
        for dist_min in K_min_dist:
            for i in range(len(list_dist)):
                if dist_min == list_dist[i]:
                    idx = i
                    index_list.append(idx)
        print(index_list)

        for idx_ in index_list:
            num_class = self.y[idx_]
            class_list.append(int(num_class))

        print(class_list) 
        
        # 5ème étape : Vote majoritaire
        class_counter = Counter(class_list)

        sorted_classes = sorted(class_counter.items(), key = lambda x: x[1], reverse = True)
        classe_majoritaire = sorted_classes[0][0]
        
        return classe_majoritaire
```

```python
self_x = X_2_data_rand_tri_gauss
self_y = Y_2_data_rand_tri_gauss

X = np.array((0.56,2.87))

knn = KNNClassifier()
knn.fit(self_x, self_y)
class_pred = knn.predict(X)

print("La classe prédite pour le point X par la KNN from scratch est : ", class_pred)
```

```python
from sklearn.metrics.pairwise import euclidean_distances

def Xtrain(X):
    Xred= [X[i] for i in range(len(X)) if i%2==0]
    return Xred
def Ytrain(Y):
    Yred= [Y[i] for i in range(len(Y)) if i%2==0]
    return Yred

def Xtest(X):
    Xt= [X[i] for i in range(len(X)) if i%2!=0]
    return Xt
def Ytest(Y):
    Yt= [Y[i] for i in range(len(Y)) if i%2!=0]
    return Yt


n_neighbors=3

x_train_knn = Xtrain(X_2_data_rand_tri_gauss)
y_train_knn = Ytrain(Y_2_data_rand_tri_gauss)
x_test_knn = Xtest(X_2_data_rand_tri_gauss)


list_dist = euclidean_distances(x_train_knn, x_test_knn)

index_dist_ = np.argsort(list_dist, axis=1)
print(index_dist_)

index_dist_min = index_dist_[:, :n_neighbors]
print(index_dist_min)

class_list = list()
for i in range(len(index_dist_min)):
    class_list_temp = list()
    for idx in index_dist_min[i]: 
        num_class = y_train_knn[idx]
        class_list_temp.append(int(num_class))
    class_list.append(class_list_temp)
    
print(class_list) 
        
# 5ème étape : Vote majoritaire
class_list_majoritaire = list()

for i in range(len(class_list)):
    class_counter = Counter(class_list[i])
    sorted_classes = sorted(class_counter.items(), key = lambda x: x[1], reverse = True)
    print(sorted_classes[0][0])
    classe_majoritaire = sorted_classes[0][0]
class_list_majoritaire.append(classe_majoritaire)
print(class_list_majoritaire)
```
